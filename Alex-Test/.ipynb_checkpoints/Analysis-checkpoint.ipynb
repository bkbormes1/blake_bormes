```python
import pandas as pd
from pyspark.sql.functions import explode, split
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType
import warnings
```

```python
df = spark.read.parquet('/tmp/games1')
df.registerTempTable('games')
```

```python
pdf = df.toPandas()
pdf = pdf.dropna()
pdf
```

```python
pdf["TotalScore"] = pdf["HomeScore"] + pdf["AwayScore"]
pdf["ScoreDiff"] = pdf["TotalScore"] - pdf["OverUnder"]
pdf
```

```python
# How many games were OVER the total bet line
pdf["ScoreDiff"].pct_change().fillna(0).gt(0).sum()
```

```python
# How many games were UNDER the total bet line
pdf["ScoreDiff"].pct_change().fillna(0).lt(0).sum()
```

```python
def week(number):
    week_seas = pdf[pdf["Week"] == number]
    return week_seas
```

```python
week("3")
```

```python
## Filter For Team
def season(team):
    team_seas = pdf[(pdf["AwayTeam"] == team) | (pdf["HomeTeam"] == team)]
    return team_seas
```

```python
TEN = season("TEN")
TEN
```
